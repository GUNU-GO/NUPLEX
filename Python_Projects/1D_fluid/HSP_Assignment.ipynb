```python
from scipy.integrate import odeint
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import unit_impulse
```

```python
def step_func(t, point):
    if t >= point:
        return 1
    else:
        return 0
    
def func1(z, t):
    dzdt = [z[1], -3*z[1]-2*z[0]-step_func(t, 1)-step_func(t, 2)]
    return dzdt

def func2(z, t):
    dzdt = [z[1], -z[0]+np.sin(2*t)*step_func(t, np.pi/2)]
    return dzdt

def func3(z, t):
    dzdt = [z[1], -4*z[0]+np.cos(2*t)*step_func(t, np.pi/2)]
    return dzdt

def func4(z, t):
    dzdt = [z[1], -4*z[0]+delta_function(t,np.pi)+delta_function(t,2*np.pi)]
    return dzdt

def delta_function(t, point):
    if t == point:
        return 1
    else:
        return 0
    
y0 = [0,0]
t_range = np.linspace(0,100,1000)
t4_range = np.linspace(0,1000,1000)

sol1 = odeint(func1, y0, t_range)
sol2 = odeint(func2, y0, t_range)
sol3 = odeint(func3, y0, t_range)
sol4 = odeint(func4, y0, t_range)
```

```python
plt.plot(t_range, sol1[:,0])
plt.legend(["$y^{\prime\prime}+3y^{\prime}+2y=u(t-1)-u(t-2)$"])
plt.ylabel('y')
plt.xlabel('t')
plt.savefig('hsp_1.png')
```

```python
plt.plot(t_range, sol2[:,0],'orange')
plt.legend(["$y^{\prime\prime}+y=sin(2t)u(t-\pi/2)$"],loc='upper right')
plt.ylabel('y')
plt.xlabel('t')
plt.savefig('hsp_2.png')
```

```python
plt.plot(t_range, sol3[:,0],'r')
plt.legend(["$y^{\prime\prime}+4y=cos(2t)u(t-\pi/2)$"])
plt.ylabel('y')
plt.xlabel('t')
plt.savefig('hsp_3.png')
```

```python
plt.plot(t_range, sol4[:,0],'teal')
plt.legend(["$y^{\prime\prime}+4y=\delta(t-\pi)-\delta(t-2\pi)$"])
plt.ylabel('y')
plt.xlabel('t')
plt.savefig('hsp_4.png')
```
