```python
## calculation_noinv import 하면서 xs.py에서 EEDF(현재:Maxwellian가정) 고려한 rate_const계산
import graph
import numpy as np
import matplotlib.pyplot as plt
```

```python
import calculation_noinv
```

```python
# [Te, ne, nH, nH_2s, nH2_v0, nH2_v1, nH2_v2, nH2_v3, nH2_v4, nH2_v5, nH2_v6, nH2_v7, nH2_v8, nH2_v9, nHp, nH2p, nH3p, nHm]
#np.seterr(all='ignore')
model = calculation_noinv.Model(5,2000,0.5,2e-6,time_resolution=1e-8)
exp_condition = (model.p, model.input_power, model.duty, model.period ,model.time_resolution)
Model_Result = model.calculation()
```

```python
graph.plot(Model_Result, 'Te', exp_condition)
graph.plot(Model_Result, 'ne', exp_condition)
graph.plot(Model_Result, 'ne,nHm,nHp,nH2p,nH3p,nH', exp_condition)
```

```python
graph.plot(Model_Result, 'nH2_v0,nH2_v1,nH2_v2,nH2_v3,nH2_v4,nH2_v5,nH2_v6,nH2_v7,nH2_v8,nH2_v9', exp_condition)
```

```python
t = np.linspace(0, model.period, int(model.period/model.time_resolution))
plt.plot(t*1e6, Model_Result[-1]/Model_Result[1]) #alpha 값
plt.title("$α$")
plt.xlabel('Time(μs)')
plt.show()
```
