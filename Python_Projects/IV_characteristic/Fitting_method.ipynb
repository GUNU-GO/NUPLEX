```python
import pandas as pd
import numpy as np
from scipy.interpolate import CubicSpline
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel('C:/Users/GUNU/Dropbox/PRL/Python_Projects/IV_characteristic/IV_Characteristic_test.xlsx')
IV_data = data[['Voltage (V)', 'Current (A)']]
EEPF_data = data[['Energy (eV)', 'EEPF (eV^1.5/m^3)']]
```

```python
voltage = np.array(IV_data['Voltage (V)'])
current = np.array(IV_data['Current (A)'])
electron_energy = np.array(EEPF_data['Energy (eV)'])
probability = np.array(EEPF_data['EEPF (eV^1.5/m^3)'])
```

```python
plt.plot(voltage, current)
plt.xlabel('Current (A)')
plt.ylabel('Voltage (V)')
plt.xlim(-150, 50)
plt.ylim(-0.05, 0.2)
plt.grid(True)
plt.show()
```

```python
plt.plot(electron_energy, probability)
plt.yscale('log')
plt.xlim(0,40)
plt.grid(True)
```
