```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint
```

```python
def func(y, t):
    dydt = np.sin(t - y)
    return dydt

y0 = 1
t = np.arange(0,2*np.pi,0.1)

y = odeint(func, y0, t)

plt.plot(t, y)
```
